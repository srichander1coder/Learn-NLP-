```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
def wordcloudd(wlist):
    text = " ".join(cat.split()[0] for cat in wlist)
    word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
wordcloudd(lincoln1)
```

```python
wordcloudd(lincoln2)
```

```python
wordcloudd(obama1)
```

```python
wordcloudd(obama2)
```

```python
wordcloudd(trump)
```

```python
wordcloudd(biden)
```

```python
from nltk.corpus import names
```

```python
labelled_names= ([(name,'male') for name in names.words('male.txt')]+[(name,'female') for name in names.words('female.txt')])
```

```python
labelled_names
```

```python
import random
random.shuffle(labelled_names)
```

```python
labelled_names
```

```python
def gender_features(n):
    return n[len(n)-1]

featuresets= ([(gender_features(n),gender) for (n,gender) in labelled_names])
featuresets
```

```python
train_set, test_set= featuresets[500:],featuresets[:500]
```

```python
import nltk
c= nltk.NaiveBayesClassifier
classifier=c.train(train_set)
```

```python
classifier.classify(gender_features('Obama'))
```
